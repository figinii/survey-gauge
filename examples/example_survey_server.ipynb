## Load Questionnaire and Scenarios from YAML

Load the sample questionnaire and scenarios from the examples folder.

```python
import sys
from pathlib import Path

root_dir = Path.cwd()
if not (root_dir / "src").exists():
    root_dir = root_dir.parent

sys.path.insert(0, str(root_dir / "src"))

from survey_gauge import Questionnaire, Scenario

examples_dir = root_dir / "examples"
questionnaire = Questionnaire.from_yml(examples_dir / "capitals_questionnaire.yml")
scenarios = Scenario.from_yml(examples_dir / "capitals_scenarios.yml")

scenario = scenarios[0]
```

## Define a Dummy Prompt Client

Return a fixed choice for every prompt to simulate scoring.

```python
from typing import List


def prompt_client(prompts: List[str]) -> List[str]:
    default_choice = questionnaire.default_choices[0].choice
    print(prompts[0])
    return [default_choice for _ in prompts]
```

## Initialize SurveyServer

Create a logger and construct the SurveyServer with the dummy prompt client.

```python
import logging

from survey_gauge.SurveyServer import SurveyServer

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger("survey_gauge.example")

server = SurveyServer(prompt_client=prompt_client, questionnaire=questionnaire, logger=logger)
```

## Run evaluate_scenario and Display Scores

Evaluate the first scenario and display the returned scores.

```python
scores = await server.evaluate_scenario(scenario)
print(scores)
```
