```python
import os

from survey_gauge import Questionnaire
from survey_gauge import Scenario
from survey_gauge import Eval

import logging
from vllm import AsyncEngineArgs

%load_ext autoreload
%autoreload 2
```

# Data

```python
questionnaire = Questionnaire.from_yml('../examples/questionnaire.yml')
scenario1, scenario2 = Scenario.from_yml('../examples/scenarios.yml')
```

# Evaluation

```python
os.environ["HF_TOKEN"] = "your_huggingface_token_here"
```

```python
config = AsyncEngineArgs(
  model="meta-llama/Llama-3.1-8B-Instruct",
  gpu_memory_utilization=0.9,
  max_model_len=4096,
  dtype="auto",
  trust_remote_code=False
)
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger("Evaluation")

evaluator = Eval(config=config, questionnaire=questionnaire,logger=logger)
```

```python
await evaluator.evaluate_scenario(scenario1)
```

```python
await evaluator.evaluate_scenario(scenario2)
```
